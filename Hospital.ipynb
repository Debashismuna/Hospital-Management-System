```python
import random
import time
import datetime
from tkinter import*
from tkinter import ttk
import tkinter.messagebox as tmsg
import sqlite3

conn=sqlite3.connect("Hospita.db")

my_cursor = conn.cursor()


class Hospital:
    def __init__(self,root):
        self.root = root
        self.root.title("Hospital Management System")
        self.root.geometry("1300x800+0+0")

        self.nameoftab=StringVar()
        self.ref=StringVar()
        self.dose=StringVar()
        self.numoftab=StringVar()
        self.lot=StringVar()
        self.admissiondate=StringVar()
        self.ptaddress=StringVar()
        self.dailydose=StringVar()
        self.sideeffect=StringVar()
        self.patientid=StringVar()
        self.dateofbath=StringVar()
        self.patientname=StringVar()
        self.phnum=StringVar()
        self.prestext=StringVar()

        Labeltitle= Label(self.root,bd=20,relief=RIDGE,text="Hospital Management System",fg="red",bg="white",
                          font=("times new roman", 40, "bold"))
        Labeltitle.place(x=0,y=0,width=1300,height=110)
        dataframe= Frame(self.root,bd=20,relief=RIDGE,).place(x=0,y=120,width=1350,height=300)

        dataframeleft = LabelFrame(dataframe,bd=10,padx=20,relief=RIDGE,font=("arial",15,"bold"),
                                   text="Patient Information")
        dataframeleft.place(x=20,y=140,width=800,height=260)

        dataframeright= LabelFrame(dataframe,bd=10,relief=RIDGE,font=("arial",15,"bold"),
                                   text="Prescription").place(x=830,y=140,width=430,height=260)

        #####Buttons frame
        buttonframe = Frame(self.root, bd=10, relief=RIDGE)
        buttonframe.place(x=0, y=425, width=1300, height=55)

        #####Details frame
        Detailsframe = Frame(self.root, bd=10, relief=RIDGE)
        Detailsframe.place(x=0, y=485, width=1280, height=150)

        ######################Data frame left#########
        tablet= Label(dataframeleft,text="Name of Tablet",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=0,column=0)
        tablet= Label(dataframeleft,text="Reference No.    ",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=1,column=0)
        tablet= Label(dataframeleft,text="Dose",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=2,column=0)
        tablet= Label(dataframeleft,text="No of Tab.",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=3,column=0)
        tablet= Label(dataframeleft,text="LOT",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=4,column=0)
        tablet= Label(dataframeleft,text="Daily Dose",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=5,column=0)
        tablet= Label(dataframeleft,text="Side Effect",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=6,column=0,)
        tablet= Label(dataframeleft,text="Name of Patient",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=0,column=4,padx=70)
        tablet= Label(dataframeleft,text="Patient ID",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=1,column=4)
        tablet= Label(dataframeleft,text="Date of Birth",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=2,column=4)
        tablet= Label(dataframeleft,text="Phone No.",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=3,column=4)
        tablet= Label(dataframeleft,text="Patint Address",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=4,column=4)
        tablet= Label(dataframeleft,text="Addimission Date",font=("arial",10,"bold"),padx=2,
                      pady=6).grid(row=5,column=4)
                       pady=6).grid(row=6,column=4)


        ######################combo box##################

        comtab= ttk.Combobox(dataframeleft,textvariable=self.nameoftab,font=("arial",10,"bold"))
        comtab["values"]=("Nice","Corona Vacacine","Atavin")
        comtab.grid(row=0,column=1,padx=25,pady=10)

       ###############Entry field##########
        Entry(dataframeleft,textvariable=self.ref,font=("arial",10,"bold"),width=24).grid(row=1,column=1)
        Entry(dataframeleft,textvariable=self.dose,font=("arial",10,"bold"),width=24).grid(row=2,column=1)
        Entry(dataframeleft,textvariable=self.numoftab,font=("arial",10,"bold"),width=24).grid(row=3,column=1)
        Entry(dataframeleft,textvariable=self.lot,font=("arial",10,"bold"),width=24).grid(row=4,column=1)
        Entry(dataframeleft,textvariable=self.dailydose,font=("arial",10,"bold"),width=24).grid(row=5,column=1)
        Entry(dataframeleft,textvariable=self.sideeffect,font=("arial",10,"bold"),width=24).grid(row=6,column=1)
        Entry(dataframeleft,textvariable=self.patientname,font=("arial",10,"bold"),width=24).grid(row=0,column=5)
        Entry(dataframeleft,textvariable=self.patientid,font=("arial",10,"bold"),width=24).grid(row=1,column=5)
        Entry(dataframeleft,textvariable=self.dateofbath,font=("arial",10,"bold"),width=24).grid(row=2,column=5)
        Entry(dataframeleft,textvariable=self.phnum,font=("arial",10,"bold"),width=24).grid(row=3,column=5)
        Entry(dataframeleft,textvariable=self.ptaddress,font=("arial",10,"bold"),width=24).grid(row=4,column=5)
        Entry(dataframeleft,textvariable=self.admissiondate,font=("arial",10,"bold"),width=24).grid(row=5,column=5)
        # Entry(dataframeleft,textvariable=self.admissiondate,font=("arial",10,"bold"),width=24).grid(row=6,column=5)



        #####################Data frame Right##########

        self.txtpres= Text(dataframeright,font=("arial",10,"bold"),width=56,height=12,padx=2,
                      pady=6).place(x=845,y=175)


                #########Buttons##########
        buttonpres = Button(buttonframe,text="Presciption Data",command= self.iprescriptiondata,font=("arial",10,"bold"),bg="green",
                            fg="white",width=22,pady=2).grid(row=0,column=0)
        buttonpres = Button(buttonframe, text="Clear",command=self.clear1 ,font=("arial", 10, "bold"), bg="green",
                            fg="white", width=22,pady=2).grid(row=0, column=1)
        buttonpres = Button(buttonframe, text="Presciption",command=self.prescription ,font=("arial", 10, "bold"), bg="green",
                            fg="white", width=22,pady=2).grid(row=0, column=2)
        buttonpres = Button(buttonframe, text="Delete",command=self.idelete ,font=("arial", 10, "bold"), bg="green",
                            fg="white", width=22,pady=2).grid(row=0, column=3)
        buttonpres = Button(buttonframe, text="Update",command=self.updatedata ,font=("arial", 10, "bold"), bg="green",
                            fg="white", width=22,pady=2).grid(row=0, column=4)
        buttonpres = Button(buttonframe, text="Exit",command=self.iexit, font=("arial", 10, "bold"), bg="green",
                            fg="white", width=22,pady=2).grid(row=0, column=5)
        buttonpres = Button(buttonframe, text="Presciption", font=("arial", 10, "bold"), bg="green",
                            fg="white", width=22,pady=2).grid(row=0, column=6)


        ##########################Table#################
        ########################scrollbar###############
        scroll_x=ttk.Scrollbar(Detailsframe,orient=HORIZONTAL)
        scroll_y=ttk.Scrollbar(Detailsframe,orient=VERTICAL)

        self.table1 =ttk.Treeview(Detailsframe,columns=("nameoftab","ref","dose","numoftab",
                                                  "lot","dailydose","sideeffect","patientname","patientid",
                                                  "dateofbath","phnum","ptadress","admissiondate"),xscrollcommand=scroll_y.set,
                            yscrollcommand=scroll_x.set)
        self.table1.heading("nameoftab",text="Name Of Tablet")
        self.table1.heading("ref",text="Ref. No.")
        self.table1.heading("dose",text="Dose")
        self.table1.heading("numoftab",text="No.of Tab.")
        self.table1.heading("lot",text="Lot")
        self.table1.heading("dailydose",text="Daily Dose")
        self.table1.heading("sideeffect",text="Side Effect")
        self.table1.heading("patientname",text="Name Of Patient")
        self.table1.heading("patientid",text="ID of Patient")
        self.table1.heading("dateofbath",text="Date Of birth")
        self.table1.heading("phnum",text="phone No.")
        self.table1.heading("ptadress",text="Address")
        self.table1.heading("admissiondate",text="Date of Addmission")

        self.table1.column("nameoftab", width=30)
        self.table1.column("ref",width=30)
        self.table1.column("nameoftab", width=30)
        self.table1.column("nameoftab", width=30)
        self.table1.column("dose", width=30)
        self.table1.column("numoftab",width=30)
        self.table1.column("lot", width=30)
        self.table1.column("dailydose", width=30)
        self.table1.column("sideeffect", width=30)
        self.table1.column("patientname", width=30)
        self.table1.column("patientid", width=30)
        self.table1.column("dateofbath",width=30)
        self.table1.column("phnum", width=30)
        self.table1.column("ptadress", width=30)
        self.table1.column("admissiondate", width=30)

        self.table1["show"]="headings"


        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=LEFT,fill=Y)
        scroll_x.config(command=self.table1.xview)
        scroll_y.config(command=self.table1.yview)



        self.table1.pack(fill=BOTH,expand=1)
        self.table1.bind("<ButtonRelease-1>",self.get_value)
        self.Show()
        ############functionality#########################

    def iprescriptiondata(self):

        if self.nameoftab.get() =="" or self.patientid.get()=="" or self.ref.get()=="":
            tmsg.showerror("Error","All Fields are Required")

        else:
            conn = sqlite3.connect("Hospita.db")

            my_cursor = conn.cursor()
            my_cursor.execute("insert into table1 VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)",
                              (self.nameoftab.get(),self.ref.get(),self.dose.get(),self.numoftab.get(),
                               self.lot.get(),self.dailydose.get(),self.sideeffect.get(),
                               self.patientname.get(),self.patientid.get(),self.dateofbath.get(),self.phnum.get(),
                               self.ptaddress.get(),self.admissiondate.get()))
            conn.commit()
            self.clear1()
            self.Show()
            conn.close()

            tmsg.showinfo("success","Record has been inserted")



    def Show(self):

        my_cursor = conn.cursor()
        my_cursor.execute("select * from table1")
        rows =  my_cursor.fetchall()
        if len(rows)!=0:
            self.table1.delete(*self.table1.get_children())
            for i in rows:
                self.table1.insert("",END,values=i)

            print(rows)
            conn.commit()

    def get_value(self,bond):
        arrow=self.table1.focus()
        content= self.table1.item(arrow)
        row = content["values"]
        self.nameoftab.set(row[0])
        self.ref.set(row[1])
        self.dose.set(row[2])
        self.numoftab.set(row[3])
        self.lot.set(row[4])
        self.dailydose.set(row[5])
        self.sideeffect.set(row[6])
        self.patientname.set(row[7])
        self.patientid.set(row[8])
        self.dateofbath.set(row[9])
        self.phnum.set(row[10])
        self.ptaddress.set(row[11])
        self.admissiondate.set(row[12])


    def clear1(self):
        self.nameoftab.set("")
        self.ref.set("")
        self.dose.set("")
        self.numoftab.set("")
        self.lot.set("")
        self.dailydose.set("")
        self.sideeffect.set("")
        self.patientname.set("")
        self.patientid.set("")
        self.dateofbath.set("")
        self.phnum.set("")
        self.ptaddress.set("")
        self.admissiondate.set("")


    def prescription(self):
        self.prestext.insert(END,"Name of Tablet:\t\t\t" + self.nameoftab.get())
        self.prestext.insert(END,"Reference no:\t\t\t"+self.ref.get())
        self.prestext.insert(END,"Dose:\t\t\t"+self.dose.get())
        self.prestext.insert(END,"No.Of Tab.:\t\t\t"+self.numoftab.get())
        self.prestext.insert(END,"Lot:\t\t\t"+self.lot.get())
        self.prestext.insert(END,"Daily Dose:\t\t\t"+self.dailydose.get())
        self.prestext.insert(END,"Side Effect:\t\t\t"+self.sideeffect.get())
        self.prestext.insert(END,"Patient Name:\t\t\t"+self.patientname.get())
        self.prestext.insert(END,"Patient ID:\t\t\t"+self.patientid.get())
        self.prestext.insert(END,"Date of Barth:\t\t\t"+self.dateofbath.get())
        self.prestext.insert(END,"Phone No.:\t\t\t"+self.phnum.get())
        self.prestext.insert(END,"Patient Address:\t\t\t"+self.ptaddress.get())



    def idelete(self):

        my_cursor = conn.cursor()
        value=(self.ref.get())
        if value !="":
            query="""delete from table1 where ref=?"""
            my_cursor.execute(query,(value,))
            self.Show()
            conn.commit()
            tmsg.showinfo("Delete","Permanently Delated")
        else:
            tmsg.showerror("Error","Select a fild")


    def updatedata(self):
        my_cursor = conn.cursor()
        my_cursor.execute("""update table1 set nameoftab=?,dose=?,numoftab=?,
                                                  lot=?,dailydose=?,sideeffect=?,patientname=?,patientid=?,
                                                  dateofbath=?,phnum=?,ptaddress=?,admissiondate=?
                                                   where ref = ?""",
                          (self.nameoftab.get(),self.dose.get(), self.numoftab.get(),self.lot.get(),
                           self.dailydose.get(),self.sideeffect.get(), self.patientname.get(), self.patientid.get(),
                           self.dateofbath.get(),self.phnum.get(), self.ptaddress.get(), self.admissiondate.get(),self.ref.get()))

        conn.commit()
        self.Show()
        tmsg.showinfo("Update","Data has updated")


    def iexit(self):
        iexit= tmsg.askyesno("Hospital Management Syste","Conform to exeit")
        if iexit > 0:
            root.destroy()
            return


root=Tk()
ob = Hospital(root)
root.mainloop()

```
